```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
arthritis = np.loadtxt(fname='data/inflammation-01.csv', delimiter = ',')

```

```python
image = plt.imshow(arthritis)
plt.show()
```

```python
av_inflammation = np.mean(arthritis, axis=0)
average_plot = plt.plot(av_inflammation)
plt.show()
```

```python
max_inflam = np.max(arthritis, axis=0)
max_plt = plt.plot(max_inflam)
```

std_inflam = np.std(arthritis, axis=1)
std_plt = plt.plot(std_inflam)

```python
std_inflam = np.std(arthritis, axis=0) 
std_plt = plt.plot(std_inflam)
```

```python
figure = plt.figure(figsize=(10,3))
axes1 = figure.add_subplot(1,3,1) #1 row, 3 columns, 1st graph
axes2 = figure.add_subplot(1,3,2)
axes3 = figure.add_subplot(1,3,3)

axes1.set_ylabel('average')
axes1.plot(np.mean(arthritis, axis=0))

axes2.set_ylabel('maximum')
axes2.plot(np.max(arthritis, axis=0))

axes3.set_ylabel('minimum')
axes3.plot(np.min(arthritis, axis=0))
figure.tight_layout()
plt.show()
```
